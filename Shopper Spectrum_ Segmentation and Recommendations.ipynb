```python
#Step 4 :  Clustering Methodology:
#1 Feature Engineering:
# Calculate Recency = Latest purchase date in dataset  Customers last purchase date
latest_date = df['InvoiceDate'].max()
recency_df = df.groupby('CustomerID')['InvoiceDate'].max().reset_index()
recency_df['Recency'] = (latest_date - recency_df['InvoiceDate']).dt.days
recency_df = recency_df[['CustomerID','Recency']]
recency_df.head(10)
```

```python
# Calculate Frequency = Number of transactions per customer
frequency_df = df.groupby('CustomerID')['InvoiceNo'].nunique().reset_index()
frequency_df.rename(columns={'InvoiceNo':'Frequency'}, inplace=True)
frequency_df.head(10)
```

```python
# Calculate Monetary = Total amount spent by customer
monetary_df = df.groupby('CustomerID')['TotalAmount'].sum().reset_index()
monetary_df.rename(columns={'TotalAmount':'Monetary'}, inplace=True)
monetary_df.head(10)
```

```python
rfm_df = recency_df.merge(frequency_df, on='CustomerID').merge(monetary_df, on='CustomerID')
rfm_df.head(10)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
rfm_standardized = rfm_df.copy()
rfm_standardized[['Recency','Frequency','Monetary']] = ss.fit_transform(rfm_standardized[['Recency','Frequency','Monetary']])

rfm_standardized.head(10)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.boxplot(data=rfm_df[['Recency','Frequency','Monetary']])
plt.title("Before Scaling (Original RFM)")

plt.subplot(1,2,2)
sns.boxplot(data=rfm_standardized[['Recency','Frequency','Monetary']])
plt.title("After Scaling (Standardized RFM)")

plt.show()
```

```python
#3 Choose Clustering Algorithm (KMeans, DBScan, Hierarchial etc)
#4  Use Elbow Method , Silhouette Score to decide the number of clusters
#Elbow curve for cluster selection
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X = rfm_standardized[['Recency', 'Frequency', 'Monetary']] 

sse = []  
silhouette_scores = []    

for k in range(2,11):
    kmeans = KMeans(n_clusters=k, random_state=0, n_init='auto')
    kmeans.fit(X)
    
    sse.append(kmeans.inertia_)
    
    score = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(score)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(range(2,11), sse, 'bo--')
plt.xlabel("Number of clusters (k)")
plt.ylabel("SSE (Inertia)")
plt.title("Elbow Method for Optimal k")

plt.subplot(1,2,2)
plt.plot(range(2,11), silhouette_scores, 'ro--')
plt.xlabel("Number of clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score vs k")

plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt

X = rfm_standardized[['Recency','Frequency','Monetary']]
neighbors = NearestNeighbors(n_neighbors=6)
neighbors_fit = neighbors.fit(X)
distances, indices = neighbors_fit.kneighbors(X)

distances = np.sort(distances[:, 5])
plt.plot(distances)
plt.ylabel("k-distance")
plt.xlabel("Points sorted by distance")
plt.title("k-distance Graph for Choosing eps")
plt.yticks(np.arange(0, 21, 1))
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

X = rfm_standardized[['Recency','Frequency','Monetary']]

eps_values = np.arange(0.1, 2.0, 0.05)
sil_scores = []

for eps in eps_values:
    dbscan = DBSCAN(eps=eps, min_samples=6)
    labels = dbscan.fit_predict(X)
    
    unique_labels = set(labels)
    n_clusters = len(unique_labels) - (1 if -1 in labels else 0)
    
    if n_clusters >= 2:
        score = silhouette_score(X, labels)
        sil_scores.append(score)
    else:
        sil_scores.append(np.nan)  

plt.plot(eps_values, sil_scores, marker='o')
plt.title("Silhouette Score vs eps for DBSCAN")
plt.xlabel("eps value")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()
```

```python
# Recommendation System Approach:
# Use Item-based Collaborative Filtering
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

customer_product_matrix = df.pivot_table(
    index='CustomerID',
    columns='Description',
    values='Quantity',
    aggfunc='sum',
    fill_value=0
)
customer_product_matrix
```

```python
# Compute cosine similarity (or another similarity metric) between products based on purchase history (CustomerIDDescription matrix)
product_similarity = cosine_similarity(customer_product_matrix.T)
product_similarity_df = pd.DataFrame(
    product_similarity,
    index=customer_product_matrix.columns,
    columns=customer_product_matrix.columns
)
product_similarity_df
```

```python
import joblib
joblib.dump(product_similarity_df, "product_similarity.pkl")

print(" Saved successfully!")
```

```python
# Return top 5 similar products to the entered product name
def recommend_products(product_name, top_n=5):
    if product_name not in product_similarity_df.columns:
        return f"{product_name} not found!"
    return product_similarity_df[product_name].sort_values(ascending=False)[1:top_n+1]

recommend_products("ZINC WIRE KITCHEN ORGANISER")
```

```python
import pandas as pd
from sklearn.metrics.pairwise import euclidean_distances

customer_product_matrix2 = df.pivot_table(
    index='CustomerID',
    columns='Description',
    values='Quantity',
    aggfunc='sum',
    fill_value=0
)
customer_product_matrix2
```

```python
product_distance = euclidean_distances(customer_product_matrix2.T)
product_distance_df = pd.DataFrame(
    product_distance,
    index=customer_product_matrix2.columns,
    columns=customer_product_matrix2.columns
)
product_distance_df
```

```python
def recommend_products_euclidean(product_name, top_n=5):
    if product_name not in product_distance_df.columns:
        return f"{product_name} not found!"
    return product_distance_df[product_name].sort_values(ascending=True)[1:top_n+1]

recommend_products_euclidean("ZINC WIRE KITCHEN ORGANISER")
```
